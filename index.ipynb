## Employees and their Office (a One-to-One join)

Return a list of all of the employees with their first name, last name and the city and state of the office that they work out of (if they have one). Include all employees and order them by their first name, then their last name.

```python
#Your code here
import sqlite3
import pandas as pd
conn = sqlite3.connect('data.sqlite')
cursor = conn.cursor()

df = pd.DataFrame(cursor.execute('''
                SELECT firstName, lastName, city, state
                FROM employees
                JOIN offices
                ON employees.officeCode = offices.officeCode
                ORDER BY firstName, lastName''').fetchall())
df.columns = [i[0] for i in cursor.description]
df
```

## Customers and their Orders (a One-to-Many join)

Return a list of all of the customers. For each customer return a record for each of their order numbers, order dates and statuses.

```python
# Your code here
df = pd.DataFrame(cursor.execute('''
                            SELECT customerName,orderNumber, status
                            FROM customers
                            JOIN orders
                            ON customers.customerNumber = orders.customerNumber;''').fetchall())
df.columns = [i[0] for i in cursor.description]
df
```

## Orders and their Order Details (another One-to-Many join)

Return a list of orders. For each order return a record for each order detail within the order.

```python
# Your code here
df = pd.DataFrame(cursor.execute('''
                            SELECT orders.orderNumber, productCode,quantityOrdered, priceEach,orderLineNumber
                            FROM orders
                            JOIN orderdetails
                            ON orders.orderNumber = orderdetails.orderNumber;''').fetchall())
df.columns = [i[0] for i in cursor.description]
df
```

## Orders, Order details and Product Details (a Many-to-Many Join)

Return a list of the orders with the the order information, and all the product information for each product in the order. 

```python
# Your code here
df = pd.DataFrame(cursor.execute('''
                                    SELECT orders.orderNumber, quantityOrdered,priceEach,orderLineNumber, productName,
                                    productLine,productScale, productVendor, productDescription,quantityInStock,
                                    buyPrice,MSRP
                                    FROM orders
                                    JOIN orderDetails
                                    ON orders.orderNumber = orderdetails.orderNumber
                                    JOIN products
                                    ON orderDetails.productCode = products.productCode''').fetchall())

df.columns = [i[0] for i in cursor.description]
df
```
